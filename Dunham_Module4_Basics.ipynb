Looking for the words "mattress," "cabinet," and "wrench" on the Craigslist Grage & Moving Sales page.

```python
keepers = []
#keepers = {}
words = ['mattress', 'cabinet', 'wrench']
for eachtitle in records.keys():
    craigslist_garage = requests.get(records[eachtitle])
    craigslist_responses = BeautifulSoup(craigslist_garage.text, 'html.parser')
    sleep(5)
    post_description = craigslist_responses.find(id="postingbody")
    if any(x in post_description.text for x in words):
        print(eachtitle)
        keepers.append(eachtitle +':'+ records[eachtitle]) #keepers[eachtitle] = records[eachtitle] for the dictionary
    else:
        print('no')
```
